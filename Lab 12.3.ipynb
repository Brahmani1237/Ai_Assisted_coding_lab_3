TASK 1:  Write python code for linear_search() function to search a value
in a list and extract it’s index.

PROMPT:Generate a python program for linear_search() function to search a value in a list and extract it’s index.

```python
def linear_search(data, value):
  """
  Performs a linear search on a list to find the index of a value.

  Args:
    data: The list to search within.
    value: The value to search for.

  Returns:
    The index of the value if found, otherwise -1.
  """
  for index, item in enumerate(data):
    if item == value:
      return index
  return -1
```

Now let's get some input from you to test the function.

```python
# Get the list from the user
list_input = input("Enter a list of numbers separated by spaces: ")
data_list = [int(x) for x in list_input.split()]

# Get the value to search for from the user
search_value = int(input("Enter the value to search for: "))

# Perform the linear search
index = linear_search(data_list, search_value)

# Display the result
if index != -1:
  print(f"The value {search_value} was found at index {index}.")
else:
  print(f"The value {search_value} was not found in the list.")
```

EXPLANATION: Certainly! The code performs a linear search. It takes a list of numbers and a value to search for as input from the user. It then uses the linear_search function to find the index of the value in the list. Finally, it prints whether the value was found and, if so, its index.

TASK 2:Ask AI to implement Bubble Sort and check sorted output

PROMPT: Generate a python program to implement Bubble Sort and check sorted output

```python
def bubble_sort(data):
  """
  Sorts a list using the Bubble Sort algorithm.

  Args:
    data: The list to be sorted.

  Returns:
    The sorted list.
  """
  n = len(data)
  for i in range(n):
    # Last i elements are already in place
    for j in range(0, n - i - 1):
      # Traverse the list from 0 to n-i-1
      # Swap if the element found is greater than the next element
      if data[j] > data[j + 1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data
```

Now, let's get some input from you to test the Bubble Sort function.

```python
# Get the list from the user
list_input = input("Enter a list of numbers separated by spaces: ")
data_list = [int(x) for x in list_input.split()]

# Perform Bubble Sort
sorted_list = bubble_sort(data_list)

# Display the sorted list
print("Sorted list:", sorted_list)
```

EXPLANATION: Certainly! This code implements the Bubble Sort algorithm. It repeatedly steps through the list, compares adjacent elements, and swaps them if they are in the wrong order. This process is repeated until the list is sorted. The code also includes input and output to allow you to test the function with your own list of numbers.

PROMPT:generate a python program to solve the below case study using linear optimization:
Consider a chocolate manufacturing company that produces only two types of chocolate i.e. A and B. Both the chocolates require Milk and Choco only. Venkataramana
To manufacture each unit of A and B, the following quantities ar required:
Each unit of A requires 1 unit of Milk and 3 units of Choco
Each unit of B requires 1 unit of Milk and 2 units of Choco
 The company kitchen has a total of 5 units of Milk and 12 units of Choco. On each sale, the company makes a profit of Rs 6 per unit A sold and Rs 5 per unit B sold.
Now, the company wishes to maximize its profit. How many units of A and B shoud it produce respectively?

```python
# Install PuLP if you haven't already
!pip install pulp

# Import the library
from pulp import *

# Create a LP problem
prob = LpProblem("Chocolate Manufacturing", LpMaximize)

# Define decision variables
# x is the number of units of chocolate A
# y is the number of units of chocolate B
x = LpVariable("A", 0, None, LpInteger) # Non-negative integer variable
y = LpVariable("B", 0, None, LpInteger) # Non-negative integer variable

# Define the objective function (maximize profit)
# Profit = 6*x + 5*y
prob += 6 * x + 5 * y, "Total Profit"

# Define the constraints
# Milk constraint: 1*x + 1*y <= 5
prob += x + y <= 5, "Milk Constraint"

# Choco constraint: 3*x + 2*y <= 12
prob += 3 * x + 2 * y <= 12, "Choco Constraint"

# Solve the problem
prob.solve()

# Print the results
print("Status:", LpStatus[prob.status])
print("Optimal number of units of A:", x.varValue)
print("Optimal number of units of B:", y.varValue)
print("Maximum Profit:", value(prob.objective))
```

EXPLANATION:Certainly! This code block summarizes the results of the linear programming problem. It restates the problem, presents the calculated optimal value of the objective function (Z), and clearly states the optimal values for the variables x1 and x2 that achieve this maximum Z. Finally, it provides a brief interpretation of these optimal values in the context of the case study, explaining that these values represent the best way to maximize the objective while respecting all the given limitations.



TASK 4:Write python code to find value of x at which the function
f(x)=2X3+4x+5 will be minimum


PROMPT: Generate a python program to find value of x at which the function f(x)=2X3+4x+5 will be minimum

```python
from scipy.optimize import minimize

# Define the function
def f(x):
  return 2*x**3 + 4*x + 5

# Use minimize to find the minimum value of x
# We need to provide an initial guess for x
initial_guess = 0

result = minimize(f, initial_guess)

# Extract the optimal x value and the minimum function value
optimal_x = result.x[0]
minimum_value = result.fun

print(f"The function f(x) = 2x^3 + 4x + 5 is minimized at x = {optimal_x}"
print(f"The minimum value of the function is: {minimum_value}")
```

EXPLANATION: Certainly! This code uses the scipy.optimize.minimize function to find the value of 'x' that results in the lowest value of the function f(x) = 2x³ + 4x + 5. It starts with an initial guess for 'x' and iteratively searches for the 'x' that minimizes the function. The output shows the optimal 'x' value and the corresponding minimum value of the function.
