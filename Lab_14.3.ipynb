Task Description #3 – JavaScript Interactivity
Task: Prompt AI to generate a JS script that validates a simple login
form (non-empty username/password).


Prompt: generate a JS script that validates a simple login
form (non-empty username/password).


```python
html_content = """
<!DOCTYPE html>
<html>
<head>
<title>Login Form Validation</title>
<style>
  .error { color: red; }
</style>
</head>
<body>

<h2>Login</h2>

<form id="loginForm" onsubmit="return validateForm()">
  <div>
    <label for="username">Username:</label>
    <input type="text" id="username" name="username">
    <span id="usernameError" class="error"></span>
  </div>
  <br>
  <div>
    <label for="password">Password:</label>
    <input type="password" id="password" name="password">
    <span id="passwordError" class="error"></span>
  </div>
  <br>
  <button type="submit">Login</button>
</form>

<script>
function validateForm() {
  let username = document.getElementById("username").value;
  let password = document.getElementById("password").value;
  let usernameError = document.getElementById("usernameError");
  let passwordError = document.getElementById("passwordError");
  let isValid = true;

  usernameError.textContent = ""; // Clear previous error message
  passwordError.textContent = ""; // Clear previous error message

  if (username == "") {
    usernameError.textContent = "Username must be filled out";
    isValid = false;
  }
  if (password == "") {
    passwordError.textContent = "Password must be filled out";
    isValid = false;
  }
  return isValid; // Prevent form submission if not valid
}
</script>

</body>
</html>
"""

from IPython.display import HTML
display(HTML(html_content))
```

Explanation:This code generates an HTML page with a simple login form and includes a JavaScript function to validate the form inputs.

Here's a breakdown:

html_content = """...""": This multi-line string contains the HTML structure for the login page.
It includes a basic HTML document structure.
A <title> for the browser tab.
<style> tags within the <head> define a CSS class .error to style error messages in red.
The <body> contains:
A heading "Login".
A <form> with the ID loginForm and an onsubmit attribute that calls the validateForm() JavaScript function when the form is submitted. The return validateForm() part is important because if validateForm() returns false, the form submission will be prevented.
Input fields (<input type="text"> and <input type="password">) for username and password, each with a corresponding <label> and a <span> element with a class of error to display potential error messages.
A submit button (<button type="submit">).
A <script> section containing the JavaScript code.
<script> ... </script>: This section contains the JavaScript function validateForm().
let username = document.getElementById("username").value; and let password = document.getElementById("password").value;: These lines get the values entered by the user in the username and password input fields using their IDs.
let usernameError = document.getElementById("usernameError"); and let passwordError = document.getElementById("passwordError");: These lines get the <span> elements where error messages will be displayed, using their IDs.
let isValid = true;: Initializes a boolean variable isValid to true. This variable will track whether the form is valid.
usernameError.textContent = ""; and passwordError.textContent = "";: These lines clear any previous error messages that might have been displayed.
if (username == "") { ... }: This checks if the username variable is empty. If it is, it sets the text content of the usernameError span to "Username must be filled out" and sets isValid to false.
if (password == "") { ... }: This checks if the password variable is empty. If it is, it sets the text content of the passwordError span to "Password must be filled out" and sets isValid to false.
return isValid;: This line returns the value of the isValid variable. If isValid is false, the onsubmit attribute in the form will prevent the form from actually submitting.
from IPython.display import HTML and display(HTML(html_content)): These lines, as before, import the necessary class and display the generated HTML content in the notebook output.
In summary, this code creates a simple login form and uses JavaScript to check if the username and password fields are empty when the form is submitted. If either field is empty, it displays an error message next to the field and prevents the form from submitting.

```python
# Install Flask
!pip install Flask

# Create the Flask app
from flask import Flask, request, render_template_string

app = Flask(__name__)

# HTML for the login form
login_form_html = """
<!doctype html>
<html>
<head><title>Login</title></head>
<body>
  <h2>Login</h2>
  <form method="post" action="/login">
    Username: <input type="text" name="username"><br>
    Password: <input type="password" name="password"><br>
    <input type="submit" value="Login">
  </form>
</body>
</html>
"""

# HTML for the welcome page
welcome_html = """
<!doctype html>
<html>
<head><title>Welcome</title></head>
<body>
  <h2>Welcome, {{ username }}!</h2>
</body>
</html>
"""

# HTML for the invalid login page
invalid_html = """
<!doctype html>
<html>
<head><title>Invalid Login</title></head>
<body>
  <h2>Invalid username or password.</h2>
  <p><a href="/">Try again</a></p>
</body>
</html>
"""

# Route for the login page
@app.route('/')
def index():
  return render_template_string(login_form_html)

# Route to handle login submission
@app.route('/login', methods=['POST'])
def login():
  username = request.form['username']
  password = request.form['password']

  # Simple validation (replace with your actual authentication logic)
  if username == 'student' and password == 'password123':
    return render_template_string(welcome_html, username=username)
  else:
    return render_template_string(invalid_html)

# To run the Flask app in Colab, you need to use a tool like ngrok
# to expose the local server to the internet.
# Alternatively, you can run this code in a local environment.

# For running in Colab:
# from flask_ngrok import run_with_ngrok
# run_with_ngrok(app) # Start ngrok when app is run

# if __name__ == '__main__':
#   # For running locally:
#   # app.run(debug=True)
#   # For running in Colab (with ngrok):
#   app.run() # When using run_with_ngrok, just call app.run()
```
