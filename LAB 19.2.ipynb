```python
# Bubble Sort - Correct Python Translation of the given Java code

def bubble_sort(arr):
    n = len(arr)

    print("Original list:", arr)

    # Bubble Sort Logic (same as Java)
    for i in range(n - 1):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                temp = arr[j]
                arr[j] = arr[j + 1]
                arr[j + 1] = temp

    print("Sorted list:  ", arr)


# Test the function
numbers = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(numbers)
```

```python
# Bubble Sort with Error Handling (Task 2)

def bubble_sort_with_validation(arr):
    # ✅ Check if list is empty
    if not arr:
        print("Error: The input list is empty. Cannot perform sorting.")
        return

    # ✅ Check if list contains ONLY numeric values
    for item in arr:
        if not isinstance(item, (int, float)):
            print("Error: List contains non-numeric values. Sorting not allowed.")
            return

    n = len(arr)
    print("Original list:", arr)

    # Bubble Sort Logic
    for i in range(n - 1):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    print("Sorted list:  ", arr)


# Test Cases
bubble_sort_with_validation([64, 34, 25, 12, 22, 11, 90])
bubble_sort_with_validation([])          # Empty list test
bubble_sort_with_validation([1, "A", 3]) # Non-numeric test
```

Is the list empty? If yes, it prints an error and stops.
Does it contain only numbers? If it finds any non-numeric values (including booleans, as per our agreement), it prints an error and stops.
If these checks pass, it proceeds with the classic Bubble Sort algorithm: it repeatedly steps through the list, compares adjacent elements, and swaps them if they are in the wrong order, until the entire list is sorted.

```python
%%writefile fileio.js
const fs = require('fs');
const inputFile = 'input.txt';
const outputFile = 'output.txt';

// Ensure the input file exists or create it
if (!fs.existsSync(inputFile)) {
    try {
        fs.writeFileSync(inputFile, 'This is a default input.', 'utf8');
        console.log(Created default input file: ${inputFile});
    } catch (writeErr) {
        console.error(Error creating input file ${inputFile}:, writeErr);
        // Exit if we can't even create the input file
        process.exit(1);
    }
}


fs.readFile(inputFile, 'utf8', (err, data) => {
    if (err) {
        if (err.code === 'ENOENT') {
            console.error(Error: Input file not found at ${inputFile});
        } else if (err.code === 'EACCES') {
            console.error(Error: Permission denied to read file ${inputFile});
        } else {
            console.error("Error reading input file:", err);
        }
        return; // Stop execution if file reading fails
    }

    // Process the content (e.g., convert to uppercase)
    const processedContent = data.toUpperCase();

    fs.writeFile(outputFile, processedContent, 'utf8', (err) => {
        if (err) {
            if (err.code === 'EACCES') {
                console.error(Error: Permission denied to write to file ${outputFile});
            } else {
                console.error("Error writing output file:", err);
            }
            return; // Stop execution if file writing fails
        }
        console.log("File processed successfully.");
    });
});
```

Reads input.txt: It uses Node.js's file system module to read content from input.txt.
Processes Data: It converts the read text to uppercase.
Writes to output.txt: It then writes this uppercase text to a new file called output.txt.
Error Handling: It also includes checks to catch errors if the input file is missing or if there are issues writing to the output file.

PROMPT:Generate a code for API Call Translation Your team developed a prototype in Python to fetch weather data from an API, but the production environment only supports Java. TASK 1: Generate a code to Translate the Python script (that makes an API call and prints the response) into Java using AI-assisted coding. Ensure equivalent functionality. TASK 2:Generate a code to Add proper error handling in the Java version for cases such as invalid API key, request timeout, or no internet connection.



```python
%%writefile ApiCall.java
import java.net.URI;
import java.net.http.HttpClient;
import java.net.http.HttpRequest;
import java.net.http.HttpResponse;

public class ApiCall {
    public static void main(String[] args) {
        HttpClient client = HttpClient.newHttpClient();
        String url = "https://jsonplaceholder.typicode.com/posts/1";

        HttpRequest request = HttpRequest.newBuilder()
                .uri(URI.create(url))
                .build();

        try {
            HttpResponse<String> response = client.send(request, HttpResponse.BodyHandlers.ofString());

            System.out.println("API Response:");
            System.out.println(response.body());

        } catch (Exception e) {
            System.err.println("Error making API call: " + e.getMessage());
            e.printStackTrace();
        }
    }
}
```

EXPLANATION:This Java code defines an ApiCall class that demonstrates how to make an HTTP GET request to a specified URL (in this case, https://jsonplaceholder.typicode.com/posts/1) and handle various potential errors. It uses the java.net.http package, which is Java's modern HTTP Client.

Here's a breakdown:

HttpClient Configuration: It initializes an HttpClient with a connectTimeout of 10 seconds. This is the maximum time allowed to establish a connection to the server.
HttpRequest Setup: An HttpRequest is built for a GET request to the example URL. It also sets a timeout of 15 seconds for the entire request, meaning if the server doesn't respond within this time, a timeout error will occur.
Making the Call: The client.send() method sends the request and retrieves the HttpResponse.
Success and Error Handling: The code includes robust try-catch blocks to manage different types of exceptions:
HTTP Status Code Check: After receiving a response, it first checks if the HTTP status code is 400 or greater (indicating a client or server error, like an invalid API key if one were used). If so, it prints an error message and the response body.
ConnectException: Catches issues where the client cannot establish a connection to the server (e.g., no internet connection, server offline).
UnknownHostException: Catches errors if the hostname in the URL cannot be resolved (e.g., a typo in the domain name).
HttpTimeoutException: Catches errors if the request exceeds the configured timeout duration.
IOException: Catches general input/output errors that might occur during the request or response handling.
InterruptedException: Catches cases where the thread making the API call is interrupted.
General Exception: A catch-all for any other unexpected errors.
In essence, this code provides a resilient way to fetch data from a web API in Java, ensuring that common network and API-related issues are handled gracefully.
