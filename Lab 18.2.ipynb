TASK 1:  Connect to a Movie Database API (e.g., OMDb or TMDB) to
fetch details of a movie

PROMPT:Generate a Python script that connects to a Movie Database API (OMDb or TMDB).
The script should:

Take a movie title as input

Query the API using an API key

Handle errors (invalid movie name, missing/expired API key, request timeout)

Display the movie Title, Release Year, Genre, IMDb Rating, and Director in a clean format

Use requests library and include comments explaining the code.

```python
import requests
import os

def get_movie_details(title, api_key):
    """Fetches movie details from OMDb API."""
    base_url = "http://www.omdbapi.com/"
    params = {
        't': title,
        'apikey': api_key
    }

    try:
        response = requests.get(base_url, params=params, timeout=10)
        response.raise_for_status()  # Raise an HTTPError for bad responses (4xx or 5xx)
        data = response.json()

        if data.get('Response') == 'True':
            print(f"Title: {data.get('Title')}")
            print(f"Year: {data.get('Year')}")
            print(f"Genre: {data.get('Genre')}")
            print(f"IMDb Rating: {data.get('imdbRating')}")
            print(f"Director: {data.get('Director')}")
        else:
            print(f"Error: {data.get('Error')}")

    except requests.exceptions.Timeout:
        print("Error: The request to OMDb API timed out.")
    except requests.exceptions.RequestException as e:
        print(f"Error: Could not connect to OMDb API. {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Replace with your actual OMDb API key
# It's recommended to store your API key securely, e.g., using environment variables
# api_key = os.environ.get("OMDB_API_KEY")
# If using Colab secrets, you can use:
# from google.colab import userdata
# api_key = userdata.get('OMDB_API_KEY')

# For demonstration purposes, you can directly assign the key here
api_key = " 7a7b0968" # Replace with your actual API key

if api_key == "YOUR_API_KEY":
    print("Please replace 'YOUR_API_KEY' with your actual OMDb API key.")
else:
    movie_title = input("Enter the movie title: ")
    get_movie_details(movie_title, api_key)
```

Finance) to fetch daily stock prices

PROMPT:Generate a Python function that connects to a Stock Market API (Alpha Vantage or Yahoo Finance).
Requirements:

Ask the user to enter a stock ticker symbol (e.g., AAPL, MSFT, TSLA).

Fetch the latest daily stock prices.

Display:

Opening Price

Closing Price

High

Low

Trading Volume

Handle errors including:

Invalid ticker symbol

API call rate limit exceeded

Null or malformed API response

Use the requests library (if Alpha Vantage) or yfinance (if Yahoo Finance).
Include comments explaining the code.

```python
import yfinance as yf # Library for fetching financial data from Yahoo Finance
import pandas as pd # Used for data manipulation, though yfinance returns DataFrames

def get_stock_details(ticker_symbol):
    """
    Fetches and displays the latest daily stock prices for a given ticker symbol.

    Args:
        ticker_symbol (str): The stock ticker symbol (e.g., AAPL, MSFT).
    """

    print(f"Searching for stock: '{ticker_symbol.upper()}'...")

    try:
        # Create a Ticker object for the given symbol
        stock = yf.Ticker(ticker_symbol)

        # Fetch historical data for the last 1 day. 'period="1d"' gets the most recent full day's data.
        # 'interval="1d"' ensures daily resolution.
        hist = stock.history(period="1d", interval="1d")

        # Check if the historical data is empty, which often indicates an invalid ticker or no data
        if hist.empty:
            # Try to get info to see if the ticker is fundamentally invalid
            info = stock.info
            if not info or 'regularMarketPrice' not in info: # Check if info is empty or lacks basic price data
                print(f"Error: Invalid ticker symbol '{ticker_symbol.upper()}' or no data available.")
            else:
                print(f"No historical data found for '{ticker_symbol.upper()}' for the last day. It might be a holiday or market closed.")
            return

        # Get the latest day's data (first row if sorted by date, which it usually is)
        latest_data = hist.iloc[0]

        # Extract and display the required details
        print(f"\n--- Latest Daily Prices for {ticker_symbol.upper()} ---")
        print(f"Opening Price: {latest_data.get('Open', 'N/A'):.2f}")
        print(f"Closing Price: {latest_data.get('Close', 'N/A'):.2f}")
        print(f"High: {latest_data.get('High', 'N/A'):.2f}")
        print(f"Low: {latest_data.get('Low', 'N/A'):.2f}")
        print(f"Trading Volume: {latest_data.get('Volume', 'N/A'):,}")
        print("------------------------------------------")

    except Exception as e:
        # Catch any other unexpected errors during the process
        print(f"An unexpected error occurred for '{ticker_symbol.upper()}': {e}")
        print("Please ensure the ticker symbol is correct and try again.")

# --- Example Usage ---
if __name__ == "__main__":
    # Get stock ticker symbol from user input
    ticker = input("Enter a stock ticker symbol (e.g., AAPL, MSFT, TSLA): ")

    # Call the function to get stock details
    get_stock_details(ticker)
```

EXPLANATION:   This script fetches daily stock prices for a given ticker symbol using the yfinance library, which interacts with Yahoo Finance. It handles various error scenarios and displays the data in a clear format.

TASK 4: Build a translator using a free Translation API (e.g.,
Libre Translate, Google Translate)

PROMPT:Generate a Python script that uses a free Translation API (e.g., LibreTranslate or Google Translate API alternative).
Requirements:

Ask the user to input text and choose a target language (e.g., "es" for Spanish, "fr" for French).

Send the request to the translation API.

Handle errors such as:

Invalid or unsupported language codes

API quota exceeded

Empty text input

Implement a retry mechanism: if the first request fails, retry once before showing an error.

Display both original text and translated text clearly in formatted output.
Use requests module and include comments explaining the code.

```python
import requests
import time

def translate_text(text, target_language, retries=3, delay=5):
    """Translates text using Libre Translate API with retry mechanism."""
    base_url = "https://translate.astian.org/translate" # Public Libre Translate instance

    for attempt in range(retries):
        try:
            response = requests.post(base_url, json={
                "q": text,
                "source": "auto", # Auto-detect source language
                "target": target_language,
                "format": "text"
            }, timeout=10)

            response.raise_for_status()  # Raise an HTTPError for bad responses (4xx or 5xx)
            data = response.json()

            if "translatedText" in data:
                return data["translatedText"]
            elif "error" in data:
                print(f"Attempt {attempt + 1} failed: API Error - {data['error']}")
            else:
                print(f"Attempt {attempt + 1} failed: Unexpected response format.")

        except requests.exceptions.Timeout:
            print(f"Attempt {attempt + 1} failed: Request timed out.")
        except requests.exceptions.RequestException as e:
            print(f"Attempt {attempt + 1} failed: Request error - {e}")
        except Exception as e:
            print(f"Attempt {attempt + 1} failed: An unexpected error occurred - {e}")

        if attempt < retries - 1:
            print(f"Retrying in {delay} seconds...")
            time.sleep(delay)

    return "Translation failed after multiple retries."

# Get input from user
input_text = input("Enter the text to translate: ")
if not input_text:
    print("Error: Input text cannot be empty.")
else:
    target_lang = input("Enter the target language code (e.g., es for Spanish, fr for French): ").lower()
    # Basic validation for target language (can be expanded)
    if not target_lang.isalpha() or len(target_lang) not in [2, 3]:
         print("Error: Invalid target language code format.")
    else:
        print("\nTranslating...")
        translated_text = translate_text(input_text, target_lang)

        print("\n--- Translation Result ---")
        print(f"Original Text: {input_text}")
        print(f"Translated Text: {translated_text}")
        print("------------------------")
```

EXPLANATION: This script connects to a public LibreTranslate instance to translate text.

Public LibreTranslate instances might have rate limits or be unstable. For production use, consider self-hosting LibreTranslate or using a paid, more reliable service.
This script assumes the LibreTranslate API is running at https://translate.argosopentech.com/ or similar. If you're using a different instance, update the LIBRETRANSLATE_API_URL variable.
No API key is typically required for self-hosted or some public LibreTranslate instances, but some might require one. This script does not include API key handling for LibreTranslate, as it's often not needed.
